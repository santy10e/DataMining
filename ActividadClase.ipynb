```python
import pandas as pd
datos = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning'
                 '-databases/autos/imports-85.data', encoding = 'utf-8',
                 )
```

```python
datos
```

```python
datos.isnull().sum()
```

Se observa en la pagina web del dataset que las variables tienen datos perdidos:
- price
- peak-rmp
- stroke
- bore
- num-of-doors
- normalized-losses
- horsepower


Se define los encabezados reales del dataset a fin de entender su informacion

```python
column_headers = [
    "symboling", "normalized_losses", "make", "fuel_type", "aspiration", "num_of_doors",
    "body_style", "drive_wheels", "engine_location", "wheel_base", "length", "width",
    "height", "curb_weight", "engine_type", "num_of_cylinders", "engine_size", "fuel_system",
    "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
]

datos.columns = column_headers
```

```python
datos
```

Se verifica los datos faltantes

```python
datos.eq('?').sum()
```

Se puede reemplazar el ? por NaN

```python
datos = datos.replace('?', pd.NA)
datos
```

- normalized-losses
Es posible predecir los datos faltantes por ello usar bfill

```python
datos.normalized_losses.fillna(method='bfill', inplace=True)
datos
```

# num-of-doors
Se puede aplicar llenar por moda, debido a que es lo comun ver autos del mismo numero de puestas

```python
moda = datos.num_of_doors.mode()
datos.num_of_doors.fillna(moda, inplace=True)
datos
```

# bore
Se puede suponer la media


```python
datos['bore'] = pd.to_numeric(datos['bore'], errors='coerce')  # 'coerce' convierte las cadenas no vÃ¡lidas a NaN

media = datos['bore'].mean()

print("La media de 'bore' es:", media)
datos.bore.fillna(media, inplace=True)
datos
```

# stroke     
Es posible predecir los datos faltantes por ello usar bfill, pues se encuentra en orden


```python
datos.stroke.fillna(method='bfill', inplace=True)
datos
```

# horsepower   
Es posible predecir los datos faltantes por ello usar bfill, pues se encuentra en orden


```python
datos.horsepower.fillna(method='bfill', inplace=True)
datos
```

# peak-rpm
En mi opinion es un dato importante, por ello no lo reemplzaria, entonces se elimina

```python
datos = datos.dropna(subset=['peak_rpm'])
```

# price                 

En mi opinion es un dato importante, por ello no lo reemplzaria, entonces se elimina

```python
datos = datos.dropna(subset=['price'])
```
