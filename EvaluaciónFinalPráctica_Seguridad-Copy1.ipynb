### Con las respuestas anteriores, se toma la decisión de mapear los datos para hacer una mejor clasificación de las respuestas

```python
# Eliminar el espacio al final del nombre de la columna
df.rename(columns={'Género ': 'Género'}, inplace=True)

# Mapear los valores existentes en la columna 'Género' a categorías específicas
mapeo_genero = {
    'Masculino': 'Hombre',
    'Hombre': 'Hombre',
    'hombre': 'Hombre',
    'Femenino': 'Mujer',
    'Mujer': 'Mujer',
    'mujer': 'Mujer',
    'Prefiero no decirlo': 'Prefiero no decirlo'
    # Añade más mapeos si es necesario para otros valores como 'Otro'
}


# Aplicar el mapeo a la columna 'Género' y asignar 'Otro' a los valores no mapeados
df['Género'] = df['Género'].map(lambda x: mapeo_genero.get(x, 'Otro'))

# Mostrar los resultados
print(df['Género'].value_counts())
```

### Existen preguntas que tienen múltiples opciones, por lo que se utilizará nuevamente el mapeo para poder hacer una mejor clasificación, en este caso las opciones múltiples no ayudan a realizar un buen análisis de los datos,

```python
mapeo_dispositivo = {
    'Smartphone.': 'Smartphone',
    'Smartphone.;Computadora/portátil.': 'Smartphone',
    'Computadora/portátil.': 'Computadora/portátil',
    'Otros dispositivos (especificar).': 'Otros dispositivos',
    'Smartphone.;Computadora/portátil.;Tableta.': 'Smartphone',
    'Tableta.': 'Tableta',
    'Smartphone.;Tableta.': 'Smartphone',
    'Smartphone.;Computadora/portátil.;Tableta.;Otros dispositivos (especificar).': 'Smartphone',
    'Smartphone.;Computadora/portátil.;Otros dispositivos (especificar).': 'Smartphone',
    'Computadora/portátil.;Otros dispositivos (especificar).': 'Computadora/portátil',
    'Smartphone.;Otros dispositivos (especificar).': 'Smartphone',
    'Computadora/portátil.;Tableta.': 'Computadora/portátil',
    'Computadora/portátil.;Tableta.;Otros dispositivos (especificar).': 'Computadora/portátil'
}
df['¿Con qué dispositivo accede?'] = df['¿Con qué dispositivo accede?'].map(mapeo_dispositivo)
```

```python
mapeo_redes_sociales = {
    'Facebook.': 'Facebook',
    'Tik Tok.': 'Tik Tok',
    'Snapchat.': 'Snapchat',
    'Whatsapp.': 'Whatsapp',
    'Telegram.': 'Telegram'
}

def clasificar_redes_sociales(opciones):
    for red_social, categoria in mapeo_redes_sociales.items():
        if red_social in opciones:
            return categoria
    return 'Otro'

# Aplicar la clasificación a la columna de opciones múltiples
df['¿Qué redes sociales usas regularmente?'] = df['¿Qué redes sociales usas regularmente?'].apply(lambda x: clasificar_redes_sociales(x.split(';')))
```
