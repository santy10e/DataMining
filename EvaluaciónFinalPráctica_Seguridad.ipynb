### Antes de realizar operaciones sobre los datos, quiero clasificar y contabilizar los datos, vamo a darle

```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('encuesta.csv')

# Eliminar el espacio al final del nombre de la columna
df.rename(columns={'Género ': 'Género'}, inplace=True)

# Mapear los valores existentes en la columna 'Género' a categorías específicas
mapeo_genero = {
    'Masculino': 'Hombre',
    'Hombre': 'Hombre',
    'hombre': 'Hombre',
    'Femenino': 'Mujer',
    'Mujer': 'Mujer',
    'mujer': 'Mujer',
    'Prefiero no decirlo': 'Prefiero no decirlo'
    # Añade más mapeos si es necesario para otros valores como 'Otro'
}


# Aplicar el mapeo a la columna 'Género' y asignar 'Otro' a los valores no mapeados
df['Género'] = df['Género'].map(lambda x: mapeo_genero.get(x, 'Otro'))

# Mostrar los resultados
print(df['Género'].value_counts())
```

```python
mapeo_frecuencia_redes = {
    'Nunca.': 'Nunca',
    'Raramente.': 'Raramente',
    'Ocasionalmente.': 'Ocasionalmente',
    'Frecuentemente.': 'Frecuentemente'
}

# se aplica el mateo
df['¿Con qué frecuencia accede a redes sociales?'] = df['¿Con qué frecuencia accede a redes sociales?'].replace(mapeo_frecuencia_redes)

# vemos resultados
print(df['Género'].value_counts())
print(df['¿Con qué frecuencia accede a redes sociales?'].value_counts())
```

### Ahora, seguiremos con las siguientes variables, al final generamos un nuevo dataset y podremos seguir con las operaciones, como limpieza de los datos, y ver que variable nos interesa para comenzar a realizar el informe

```python
#/MAPEO con que dispositivo accede
mapeo_dispositivo_acceso = {
    'Smartphone.': 'Smartphone',
    'Computadora/portátil.': 'Computadora/portátil',
    'Tableta.': 'Tableta',
    'Otros dispositivos (especificar).': 'Otros dispositivos'
}

df['¿Con qué dispositivo accede?'] = df['¿Con qué dispositivo accede?'].replace(mapeo_dispositivo_acceso)
#/// MApeo / Que redes sociales usa regularmente
mapeo_redes_sociales = {
    'Facebook.': 'Facebook',
    'Tik Tok.': 'Tik Tok',
    'Snapchat.': 'Snapchat',
    'Whatsapp.': 'Whatsapp',
    'Telegram.': 'Telegram'
}

df['¿Qué redes sociales usas regularmente?'] = df['¿Qué redes sociales usas regularmente?'].replace(mapeo_redes_sociales)

#// Mapeo . desde cuando utsa redes sociales
mapeo_tiempo_redes_sociales = {
    'Menos de 1 año.': 'Menos de 1 año',
    '1-2 años.': '1-2 años',
    '3-5 años.': '3-5 años',
    '6-10 años.': '6-10 años',
    'Más de 10 años.': 'Más de 10 años'
}

df['¿Desde cuándo utilizas redes sociales?'] = df['¿Desde cuándo utilizas redes sociales?'].replace(mapeo_tiempo_redes_sociales)

#//Mapeo = cuales son las principales razones para usar redes sociales
df.rename(columns={'¿Cuáles son tus principales razones para usar redes sociales? ': '¿Cuáles son tus principales razones para usar redes sociales?'}, inplace=True)
mapeo_razones_redes_sociales = {
    'Comunicación con amigos y familia.': 'Comunicación con amigos y familia',
    'Acceso a información y noticias.': 'Acceso a información y noticias',
    'Entretenimiento (videos, música, juegos)': 'Entretenimiento',
    'Educación o aprendizaje en línea.': 'Educación o aprendizaje en línea'
}

df['¿Cuáles son tus principales razones para usar redes sociales?'] = df['¿Cuáles son tus principales razones para usar redes sociales?'].replace(mapeo_razones_redes_sociales)

#// Mapeo -¿Quién puede ver tu perfil y qué tipo de información compartes?
mapeo_privacidad_perfil = {
    'Todo el mundo (perfil público).': 'Todo el mundo',
    'Solo mis amigos/conexiones.': 'Solo mis amigos/conexiones',
    'Solo personas que sigo y me siguen de vuelta.': 'Solo personas que sigo y me siguen de vuelta',
    'Solo yo (perfil privado).': 'Solo yo'
}

df['¿Quién puede ver tu perfil y qué tipo de información compartes?'] = df['¿Quién puede ver tu perfil y qué tipo de información compartes?'].replace(mapeo_privacidad_perfil)

#//Mapeo - ¿Has experimentado exposición involuntaria de información personal?
mapeo_exposicion_involuntaria = {
    'Sí': 'Sí',
    'No': 'No'
}

df['¿Has experimentado exposición involuntaria de información personal?'] = df['¿Has experimentado exposición involuntaria de información personal?'].replace(mapeo_exposicion_involuntaria)

#//Mapeo ¿Qué medidas de seguridad has tomado para proteger tu privacidad y seguridad en línea?
mapeo_medidas_seguridad = {
    'He ajustado las configuraciones de privacidad en mis perfiles de redes sociales.': 'Configuraciones de privacidad ajustadas en perfiles de redes sociales',
    'Uso configuraciones de privacidad para controlar quién puede ver mis publicaciones.': 'Uso de configuraciones de privacidad en publicaciones',
    'Realizo copias de seguridad de mis datos importantes regularmente.': 'Realización de copias de seguridad de datos importantes regularmente',
    'Ninguna': 'Ninguna'
}

df['¿Qué medidas de seguridad has tomado para proteger tu privacidad y seguridad en línea?'] = df['¿Qué medidas de seguridad has tomado para proteger tu privacidad y seguridad en línea?'].replace(mapeo_medidas_seguridad)

#//Mapeo - 
```
