```python
import numpy as np
import pandas as pd

dataset_path = 'limpieza.csv'

# Lee el archivo CSV con los ajustes
datos = pd.read_csv(dataset_path)
datos
```

Isnull detectar datos nulos

```python
datos.isnull().sum()
```

```python
datos.notnull().sum()
```

dropna elimina filas que tienen datos nulos, solo en ese momento

```python
datos.dropna()
```

```python
datos.dropna(inplace=True)
```

```python
datos
```

```python
datos.isnull().sum()
```

```python
mean_horass = datos['Horas'].mean()
mean_horass

```

```python
mean_horas = datos['Empleados'].mean()
mean_horas
```

```python
fillna imputar valores, para rellenar los nulos con valor determinado 
```

```python
datos.Salario.fillna((datos.Salario.sum()/len(datos.Salario)).round(2), inplace=True)
```

```python
media = round(datos.Salario.mean(),2)

print(media)
```

```python
datos.Salario.fillna(media, inplace=True)
datos
```

```python
datos.Horas.fillna((datos.Horas.sum()/len(datos.Horas)).round(0), inplace=True)
```

```python
datos.Promedio.fillna((datos.Promedio.sum()/len(datos.Promedio)).round(2), inplace=True)
```

```python
datos
```

```python
datos.Horas.median()
```

```python
datos.isnull().sum()
```
