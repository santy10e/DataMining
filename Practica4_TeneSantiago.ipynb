1. Leer e investigar sobre los marcos de trabajo KDD, CRISP-DM y SEMNA 

KDD (Extracción de Conocimiento en Bases de Datos): KDD es un proceso para descubrir conocimiento útil a partir de grandes volúmenes de datos. Involucra varias etapas, como selección de datos, limpieza, preprocesamiento, transformación, minería de datos, evaluación de patrones y presentación de conocimiento. El objetivo de KDD es convertir datos en bruto en información y conocimiento que se pueda utilizar. [1]

CRISP-DM (Proceso Estándar Cruzado para la Minería de Datos): CRISP-DM es un modelo de proceso ampliamente utilizado para la minería de datos y análisis. Proporciona un enfoque estructurado para guiar las tareas involucradas en un proyecto de minería de datos. El proceso CRISP-DM consta de seis fases: Comprensión del Negocio, Comprensión de los Datos, Preparación de Datos, Modelado, Evaluación e Implementación. Es un proceso iterativo, lo que permite flexibilidad y adaptación según los requisitos del proyecto. [1]

SEMMA es el acrónimo a las cinco fases: (Sample, Explore, Modify, Model, Assess) La metodología es propuesta por SAS Institute Inc, la define como: “… proceso de selección, exploración y modelamiento de grandes cantidades de datos para descubrir patrones de negocios desconocidos.[1]

¿Qué es un marco de trabajo para trabajar con datos y por qué es importante utilizarlo en un proyecto?
Un marco de trabajo proporciona pasos o fases que se deben seguir y cumplir para realizar una exploracion de los datos en base de datos, existen varios marcos de trabajos que son utilizados para casos en especifico pero los 3 principales marcos de trabajo son: Las tres metodologías dominantes para el proceso de la minería de
datos son: KDD, CRISP-DM y SEMMA.
Los marcos de trabajo se deben seguir de una forma secuencial es decir paso por paso y nos garantizaran una forma eficiente y confiable para conseguir un buen resultado en el analisis y exploracion de datos.

¿Cuáles son las etapas del marco de trabajo elegido por Usted y cómo se relacionan con el proceso de análisis de datos?
KDD es una metodologia propuesta por Fayyad y se proponen 5 fases: Selección, preprocesamiento, transformacion, mineria de datos y evaluación e implantación. Es un proceso iterativo e interactivo. 
El proceso de Análisis de Datos conlleva la recolección, transformacion, limpieza y modelado de datos para descubrir la información útil y de interés para una organización
Se relacionan en gran medida ya que los pasos o el proceso en si, se asemejan.
